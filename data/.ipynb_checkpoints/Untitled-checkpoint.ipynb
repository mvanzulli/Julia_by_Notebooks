<span style="color:blue"> This notebook will plot a polynomial function: $f(x)=2x^2-4$ </span>

```julia
f(x) = 2x^2 -4
```

The above equation is a symetric parabola along axis $y$ and it indepndent term is $-4$. The second derivative of x respect to the $x$ is $2$, therefore the first derivative increases inside the domain.

```julia
xᵥ = collect(-10:2:10) # vector x
yᵥ = similar(xᵥ)  # initialize vector y 
for (i,xᵢ) in enumerate(xᵥ) 
    yᵥ[i] = f(xᵢ)# compute f(x_i) and fill y vector
end
```

<span style="color:blue"> In the following cell the results are plotted: </span>

```julia
# load plots
using Plots
gr()

# plot f(x)
plot(xᵥ, yᵥ,
color  = :red,
legend = :top,
marker = (:octagon, 4),
label  = "f(x)",
title  = "f(x) = 2x^2 -4",
xlabel = "x",
ylabel = "y",
line   = (:dash)
)
```
